```python
import pandas
 
# load the csv
data = pandas.read_csv("nba.csv")
 
# show first 5 column
data.head()
```

```python
import seaborn as sns
```

```python
sns.distplot(data["Age"], color = "g")
```

The kde parameter is set to False, which disables the kernel density estimate plot. The KDE plot represents the smooth estimate of the underlying distribution based on the data and can be turned on or off by setting kde to True or False

```python
sns.distplot(data["Age"],bins=10,kde=False)
```

```python
# plotting lineplot
sns.lineplot( data['Age'], data['Weight'])
```

```python
sns.lineplot( data['Age'], data['Weight'], hue = data['Position'])
```

```python
sns.scatterplot(data['Age'],data['Weight'])
```

```python
sns.scatterplot( data['Age'], data['Weight'], hue =data["Position"])
```
